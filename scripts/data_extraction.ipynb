# **Importamos la data desde el Sistema**

```python
#Primero extraemos el json hasta el directorio local

from google.colab import drive
import json

drive.mount('/content/drive')


# Dfinimos el directorio

dir_data_kaggle='./drive/MyDrive/alicorp/data'

dir=open(dir_data_kaggle+'/kaggle.json')

token=json.load(dir)

# cargamos en el directorio local

json_string = json.dumps(token)
json_file = open("kaggle.json", "w")
json_file.write(json_string)
json_file.close()
```
